## 【問題1】データの取得

```python
import pandas as pd
from sklearn.datasets import load_iris
dataset = load_iris()

# 説明変数をDataFrame型変数Xに格納
X = pd.DataFrame(dataset.data, columns=["sepal_length", "sepal_width", "petal_length", "petal_width"])

# 目的変数をDataFrame型変数yに格納
y = pd.DataFrame(dataset.target, columns=["Species"])
```

## 【問題2】データの結合

```python
df = pd.concat([X, y], axis=1)
df
```

## 【問題4】irisデータセット自体の調査

アヤメのデータセット、またはフィッシャーのアヤメデータセットは、英国の統計学者で生物学者ロナルド・フィッシャーが、1936年の論文で導入した分類学で利用する線形判別分析のデータセットです。 エドガー・アンダーソンは、3種のアヤメの形態変化を定量化するためデータ収集したため、アンダーソンのアヤメデータセットと呼ばれることがあります。 3種のうちの2種は、同じ日に、同じ牧草地、同じ装置で、同じ人が、同時に採取、測定したものです。  

データセットは、Iris（Iris setosa、Iris virginica、およびIris versicolor）の3種のそれぞれの50サンプルからなります。 各サンプルから4つの特徴である、がく（萼）と花弁（花びら）の長さと幅を測定しました。

# 【問題6】図の作成

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

label = dataset.target_names
count = df["Species"].value_counts()

plt.pie(count, labels = label, startangle=90, autopct="%1.1f%%")
plt.show()
```

```python
#特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化

features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

for feature in features:
    plt.boxplot([df[df["Species"]==0][feature], df[df["Species"]==1][feature], df[df["Species"]==2][feature]])
    plt.xticks([1, 2, 3], ['setosa', 'versicolor', 'virginica'])
    plt.title(feature)
    plt.grid()
    plt.show()
```

```python
#特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

for feature in features:
    plt.violinplot([df[df["Species"]==0][feature], df[df["Species"]==1][feature], df[df["Species"]==2][feature]])
    plt.xticks([1, 2, 3], ['setosa', 'versicolor', 'virginica'])
    plt.title(feature)
    plt.grid()
    plt.show()
```

[考察]
箱ひげ図は四分位数を把握しやすいというメリットがあるのに対し、バイオリンプロットは分布データが把握できるというメリットがあります。

## 【問題7】特徴量間の関係の確認

### 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る

```python
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

#1. sepal_length &sepal_width 
plt.title("sepal_length & sepal_width")
plt.scatter(df[df['Species']==0][features[0]], df[df['Species']==0][features[1]], c = 'red', label="setosa")
plt.scatter(df[df['Species']==1][features[0]], df[df['Species']==1][features[1]], c = 'blue', label="versicolor")
plt.scatter(df[df['Species']==2][features[0]], df[df['Species']==2][features[1]], c = 'green', label="virginica")
plt.legend()
plt.show()

#2. sepal_length & petal_length 
plt.title("sepal_length & petal_length")
plt.scatter(df[df['Species']==0][features[0]], df[df['Species']==0][features[2]], c = 'red', label="setosa")
plt.scatter(df[df['Species']==1][features[0]], df[df['Species']==1][features[2]], c = 'blue', label="versicolor")
plt.scatter(df[df['Species']==2][features[0]], df[df['Species']==2][features[2]], c = 'green', label="virginica")
plt.legend()
plt.show()

#3. sepal_length & petal_width 
plt.title("sepal_length & petal_width")
plt.scatter(df[df['Species']==0][features[0]], df[df['Species']==0][features[3]], c = 'red', label="setosa")
plt.scatter(df[df['Species']==1][features[0]], df[df['Species']==1][features[3]], c = 'blue', label="versicolor")
plt.scatter(df[df['Species']==2][features[0]], df[df['Species']==2][features[3]], c = 'green', label="virginica")
plt.legend()
plt.show()

#4. sepal_width & petal_length 
plt.title("sepal_width & petal_length")
plt.scatter(df[df['Species']==0][features[1]], df[df['Species']==0][features[2]], c = 'red', label="setosa")
plt.scatter(df[df['Species']==1][features[1]], df[df['Species']==1][features[2]], c = 'blue', label="versicolor")
plt.scatter(df[df['Species']==2][features[1]], df[df['Species']==2][features[2]], c = 'green', label="virginica")
plt.legend()
plt.show()

#5. sepal_width & petal_width 
plt.title("sepal_width & petal_width")
plt.scatter(df[df['Species']==0][features[1]], df[df['Species']==0][features[3]], c = 'red', label="setosa")
plt.scatter(df[df['Species']==1][features[1]], df[df['Species']==1][features[3]], c = 'blue', label="versicolor")
plt.scatter(df[df['Species']==2][features[1]], df[df['Species']==2][features[3]], c = 'green', label="virginica")
plt.legend()
plt.show()

#6. petal_length & petal_width 
plt.title("petal_length & petal_width")
plt.scatter(df[df['Species']==0][features[2]], df[df['Species']==0][features[3]], c = 'red', label="setosa")
plt.scatter(df[df['Species']==1][features[2]], df[df['Species']==1][features[3]], c = 'blue', label="versicolor")
plt.scatter(df[df['Species']==2][features[2]], df[df['Species']==2][features[3]], c = 'green', label="virginica")
plt.legend()
plt.show()
```
