## 【問題1】乱数の作成

```python
import numpy as np
from numpy.random import *
 
mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
 
# 2次元正規乱数を500個生成
values = multivariate_normal(mu, sigma, 500)
values.shape
```

## 【問題2】散布図による可視化

```python
from matplotlib import pyplot as plt

x1 = values[:, 0]
x2 = values[:, 1]

plt.title("Scatter")
plt.scatter(x1, x2)
plt.show()
```

## 【問題3】ヒストグラムによる可視化

```python
import matplotlib.pyplot as plt
plt.title("histogram of x1")
plt.xlim(right=3)
plt.xlim(left=-6)
plt.hist(x1)
plt.show()

plt.title("histogram of x2")
plt.xlim(right=3)
plt.xlim(left=-6)
plt.hist(x2)
plt.show()
```

## 【問題4】データの追加

```python
mu = [0, -3]
sigma = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
 
# 2次元正規乱数を500個生成
values2 = multivariate_normal(mu, sigma, 500)
values2.shape
```

```python
x0 = values[:, 0]
y0 = values[:, 1]
x1 = values2[:, 0]
y1 = values2[:, 1]

plt.title("Scatter")
plt.scatter(x0, y0, label="0")
plt.scatter(x1, y1, label="1")
plt.legend()
plt.show()
```

## 【問題5】データの結合

```python
mix_value = np.concatenate((values, values2), axis=0)
mix_value.shape
```

## 【問題6】ラベル付け

```python
label_x = np.full((1,500),0)
label_y = np.full((1,500),1)
label_x_y = np.hstack((label_x, label_y))
label_x_y.shape
```

```python
new_value = np.concatenate((mix_value, label_x_y.T), axis=1)
new_value
```
