## 【問題2】コードリーディング
論文[1]に目を通した上で、上記実装のコードリーディングを行ってください。

### model.py
- モデルの定義（畳み込み槽、プーリング層、ドロップアウト等）
- モデルのインスタンス化
- モデルのコンパイル

### main.py
- trainGeneratorのインスタンス化
- unetのインスタンス化
- ModelCheckpointをインスタンス化
- keras.modelからfit_generatorを呼び出しtrainGeneratorを学習
- testGeneratorをインスタンス化
- keras.modelからpredict_generatorを呼び出し、予測する
- saveResultにより予測値を保存

### data.py
- adjustData : 画像データを数値データに変換する関数
- trainGenerator : train用画像データを拡張し、テンソル形式に変換する関数
- testGenerator : test画像を読み込んで、テンソル形式に変換する関数
- labelVisualize : ３次元に変換した上で正規化を行っている関数
- saveResult : 推定結果を保存する関数
